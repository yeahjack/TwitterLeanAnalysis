```python
import re
for i in tqdm(range(len(data_final))):
    data_final.text[i] = re.sub(r'(@\w+)|(https?\S+)', '', data_final.text[i])
for i in tqdm(range(len(data_final))):
    data_final.text[i] = re.sub(r'[^\x00-\x7f]', r' ', data_final.text[i])
```

```python
from tqdm import tqdm
import re

def remove_non_latin(text):
    text = re.sub(r'(@\w+)|(https?\S+)', r' ', text)
    return re.sub(r'[^\x00-\x7f]', r' ', text)

data_final.text = tqdm(data_final.text.apply(remove_non_latin))
```
