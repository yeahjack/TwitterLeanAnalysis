```python
import pandas as pd
import re
from tqdm import tqdm
```

```python
df = pd.read_csv('../../data/dataset/dataset_full_info.csv', encoding='utf-8', header=0)
```

```python
df = df.assign(full_text_masked='')
```

```python
df = df.assign(source_short='')
```

```python
df
```

```python
for i in tqdm(range(len(df))):
    if '(' in df.source[i]:
        df.source_short[i] = df.source[i].split('(')[0].strip()
    else:
        df.source_short[i] = df.source[i]
```

```python
for i in tqdm(range(len(df))):
    df.loc[i, 'full_text_masked'] = re.sub(re.compile(re.escape(df.loc[i, 'source_short']), re.IGNORECASE), '', df.loc[i, 'full_text'])
```

```python
df
```

```python
df.to_csv('../../data/dataset/dataset_all_info_masked.csv', index=False, encoding='utf-8')
```
