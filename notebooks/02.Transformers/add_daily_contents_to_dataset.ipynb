```python
import pandas as pd
from tqdm import tqdm
```

```python
dataset = pd.read_csv('../../data/dataset/dataset_all_info_masked.csv')
```

```python
dataset.source.value_counts().head(10)
```

```python
dataset.head(10)
```

```python
dataset['leaning'].replace('Center', 0, inplace=True)
dataset['leaning'].replace('Left', -2, inplace=True)
dataset['leaning'].replace('Right', 2, inplace=True)
dataset['leaning'].replace('Lean Left', -1, inplace=True)
dataset['leaning'].replace('Lean Right', 1, inplace=True)
```

```python
dataset['leaning'].value_counts()
```

```python
dataset.drop(dataset[(dataset.leaning == 'Mixed') | (dataset.leaning == 'Political News Media Bias Rating: Not Rated')].index, inplace=True)
```

```python
dataset
```

```python
dataset = dataset.loc[:,['full_text_masked', 'leaning']]
```

```python
dataset = dataset.reset_index()
dataset = dataset.loc[:,['full_text_masked', 'leaning']]
```

```python
dataset.columns = ['text', 'label']
```

```python
dataset['label'].value_counts()
```

# Dataset Daily

```python
dataset_daily = pd.read_csv('../../data/dataset/dataset_daily.csv', encoding='latin-1', header=None)
```

```python
dataset_daily.loc[:,6].value_counts()
```

```python
dataset_daily[dataset_daily['lang']=='en'].to_csv('../../data/dataset/dataset_daily.csv', index=False, header=False)
```

```python
dataset_daily_formatted = pd.DataFrame(dataset_daily.iloc[:,5])
```

```python
dataset_daily_formatted
```

```python
dataset_daily_formatted = dataset_daily_formatted.assign(label=3)
```

```python
dataset_daily_formatted.columns = ['text', 'label']
```

```python
dataset_daily_formatted
```

```python
dataset_daily_sampled = dataset_daily_formatted.sample(n=50000, random_state=42)
```

```python
dataset_daily_sampled = dataset_daily_sampled.reset_index()
```

```python
dataset_daily_sampled = dataset_daily_sampled.loc[:,['text', 'label']]
```

```python
dataset_daily_sampled
```

```python
import re
for i in tqdm(range(len(dataset_daily_sampled))):
    dataset_daily_sampled.text[i] = re.sub(r'(@\w+)|(https?\S+)', '', dataset_daily_sampled.text[i])
```

```python
pd.concat([dataset, dataset_daily_sampled]).to_csv('../../data/dataset/dataset_combined.csv', index=False, header=False)
```
