```python
# Define the function to be applied to each row
def process_row(row):
    # do some processing on the row
    return None if detect(row.TweetContent) != 'en' else row.TweetContent


# Create a pool of worker processes
with Pool(128) as pool:
    # Apply the function to each row in parallel
    results = pool.map(process_row, twitter_content.itertuples(index=False))

# Convert the results back into a DataFrame
processed_df = pd.DataFrame(results)
```
